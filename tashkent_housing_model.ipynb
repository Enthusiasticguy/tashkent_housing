# **Importing necessary modules**

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Enthusiasticguy/tashkent_housing/refs/heads/main/tashkent_housing.csv')
df = df.drop('location', axis = 1)
```

# **Splitting the dataset**



*   `X_train` --> train set without label y
*   `y` ---> actual label y

*   `X_num`  ---> it is gonna be used for scaling






```python
from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(df, test_size = 0.25, random_state = 36)

X_train = df.drop('price', axis = 1)
y = df['price'].copy()

X_num = X_train.drop('district', axis = 1)
```

# **Small pipeline for scaling**

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

num_pipeline = Pipeline([
    ('scaler',StandardScaler())
])
```

# **Scaling and One hot encoding**

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(X_num)
cat_attribs = ['district']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs)
])
```

# **Fitting into the pipeline**

```python
X_prepared = full_pipeline.fit_transform(X_train)
```

**Just an example**

```python
X_prepared[1100:1101]
```

# **Linear Regression importing**

```python
from sklearn.linear_model import LinearRegression
LR_model = LinearRegression()
```

```python
LR_model.fit(X_prepared,y)
```

```python
X_test = test_set.drop('price', axis = 1)
y_test = test_set['price'].copy()
X_test_prepared = full_pipeline.fit_transform(X_test)
```

```python
y_predicted = LR_model.predict(X_test_prepared)
```

# **Evaluating the accuracy of the model using squared error**

```python
from sklearn.metrics import mean_squared_error
msr = mean_squared_error(y_test,y_predicted)
print(np.sqrt(msr))
```

# **Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
RF_model = RandomForestRegressor()
RF_model.fit(X_prepared,y)
```

```python
y_predicted = RF_model.predict(X_test_prepared)
```

# **Evaluating the accuracy of the model**

```python
from sklearn.metrics import mean_squared_error
msr = mean_squared_error(y_test,y_predicted)
print(np.sqrt(msr))
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,y_predicted)
mae
```
